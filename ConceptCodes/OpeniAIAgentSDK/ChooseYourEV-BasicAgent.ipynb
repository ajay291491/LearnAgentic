### Choose Your EV - An LLM Agentic Consultant for Electric Vehicle Recommendations (Basic Agentic Version)
Purpose : This notebook is intended to create a basic agentic consultant app demonstrating the use of OpenAI Agent SDK
This will cover
- Creating an agent with specific instructions and model
- Running the agent with a user prompt and getting the response
- Using trace to understand the flow of execution and response generation (Traces logged at - https://platform.openai.com/logs?api=traces)
- Displaying the final output in a user-friendly format using Markdown

```python
# Importing required Python Libraries

import os
from dotenv import load_dotenv
from agents import Agent, Runner, trace
from IPython.display import Markdown, display
```

```python
# Loading required variables from environment

load_dotenv(override=True)
openai_key = os.getenv("OPENAI_API_KEY")
OPENAI_API_KEY = openai_key
```

```python
# Defining an agent

agent = Agent(
    name="EV Sales Assistant",
    instructions="You must check all EV benchmarks available to you and advise customer on top 3 models to choose with its 4 year resale value",
    model="gpt-4o-mini"
)
```

```python
# invoking an agent with Trace (trace available at https://platform.openai.com/logs?api=traces)

with trace("Choose Your EV"):
    result = await Runner.run(agent, "What are the top EV models available in UK under 25K budget")
    output = result.final_output
    display(Markdown(output))
```
