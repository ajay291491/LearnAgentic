### Understanding Xuanzang Journeys

This notebook is intended to prove the LLM resource technique.
Using this notebook I am trying to provide additional resource from wikipedia as a book about "Xuanzang" a famous Chinese traveler in the ancient times and then asking question related to the book content to predict future events related to the question.

```python
# Loading required python modules

from dotenv import load_dotenv
from openai import OpenAI
from pypdf import PdfReader
import gradio as gr
```

```python
# Loading required variables and initializing OpenAi SDK

load_dotenv(override=True)
openai = OpenAI()
```

```python
# Reading main books

main_reader= PdfReader("Books/Xuanzang.pdf")
book=""
for page in main_reader.pages:
    text = page.extract_text()
    if text: book += text
```

```python
# Reading supplementary books

supplement_reader= PdfReader("Books/Asoka_-_the_Buddhist_Emperor_of_India_Chapter_2.pdf")
supplement_book=""
for page in supplement_reader.pages:
    text = page.extract_text()
    if text: supplement_book += text
```

```python
# System prompts serve as the foundational instructions that dictate an AI's behavior and responses. They provide context
system_prompt
```

```python
# Creating system prompt

assistant_name = "History Advisor"
system_prompt = f"""You are a helpful assistant with name {assistant_name} for answering question related to the book content in the book {book}. \
if there are questions related to Ashoka empire then you have to answer the question based on the content of the book and supplement book {supplement_book}.\
if you do not know the answer you can say I d not have this knowledge at this time."""
```

```python
# Creating a method to answer question based on the book content
def answer_question(message):
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": message}
    ]
    response = openai.chat.completions.create(model="gpt-4o-mini", messages=messages, max_tokens=500)
    answer = response.choices[0].message.content
    return answer
```

```python
# using gradio to launch the chat & message will be passwd as input to the method answer_question
gr.ChatInterface(answer_question).launch()
```
