```python
# Importing required Modules

import os
import json
from  dotenv import load_dotenv
from openai import OpenAI, api_key
from anthropic import Anthropic
from google import genai
from IPython.display import Markdown, display
```

```python
# Loading Environments

load_dotenv(override=True)
```

```python
# Loading LLM keys
openai_key = os.getenv("OPENAI_API_KEY")
anthropic_key = os.getenv("ANTHROPIC_API_KEY")
google_key = os.getenv("GOOGLE_API_KEY")
```

```python
print(f'OpenAI Key     : {openai_key[:5]}',
      f'Anthropic key  : {anthropic_key[:5]}',
      f'google_key     : {google_key[:5]}')
```

```python
# Initializing SDK for OpenAI, Anthropic and Google

openai=OpenAI()
claude=Anthropic()
gemini = genai.Client(api_key=google_key)
```

```python
prompt = "suggest me 3 top companies in artificial intelligence which will lead in next 2 - 3 Years also will gain larger market share by  2028"
message = [{"role": "user", "content": prompt}]
```

```python
# Connecting to OpenAI API with model "gpt-5-nano"
model_name = "gpt-5-nano"
response = openai.chat.completions.create(model="gpt-4o", messages=message, max_tokens=500)
answer_openai = response.choices[0].message.content # This is a OpenAI Chat completion response object
display(Markdown((answer_openai)))
print(answer_openai)
```

```python
validation = "Please check this advise from OpanAI is right or you have better suggestion - " + prompt + ", openai provided answer as " + answer_openai
message = [{"role": "user", "content": validation}]
model_name = "claude-sonnet-4-5"
response = claude.messages.create(model=model_name, messages=message, max_tokens=1000)
answer_anthropic = response.content[0].text
display(Markdown((answer_anthropic)))
print(answer_anthropic)
```

```python
third_party_validation = "Please check the advise from OpanAI or Anthropic is better or or you have better suggestion on - " + prompt + ", openai provided answer as " + answer_openai + "Anthropic provided answer as " + answer_anthropic
message = [{"role": "user", "content": third_party_validation}]
gemini = OpenAI(api_key=google_key, base_url="https://generativelanguage.googleapis.com/v1beta/openai/")
model_name = "gemini-2.5-flash"
response = gemini.chat.completions.create(model=model_name, messages=message)
answer = response.choices[0].message.content

display(Markdown(answer))
```
