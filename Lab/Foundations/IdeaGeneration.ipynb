```python
# Importing required Modules

import os
import json
from  dotenv import load_dotenv
from openai import OpenAI
from anthropic import Anthropic
from IPython.display import Markdown, display
```

```python
# Loading Environments

load_dotenv(override=True)
```

```python
# Loading LLM keys
openai_key = os.getenv("OPENAI_API_KEY")
anthropic_key = os.getenv("ANTHROPIC_API_KEY")
google_key = os.getenv("GOOGLE_API_KEY")
```

```python
print(f'OpenAI Key     : {openai_key[:5]}',
      f'Anthropic key  : {anthropic_key[:5]}',
      f'google_key     : {google_key[:5]}')
```

```python
# Initializing SDK for OpenAI & Anthropic

openai=OpenAI()
claude=Anthropic()
```

```python
prompt = "suggest me 3 top companies in artificial intelligence which will lead in next 3 -6 months also will make impact to share market with 10% gain in 2026"
message = [{"role": "user", "content": prompt}]
```

```python
# Connecting to OpenAI "gpt-5-nano"
model_name = "gpt-5-nano"
response = openai.chat.completions.create(model="gpt-4o", messages=message, max_tokens=500)
content = response.choices[0].message.content # This is a OpenAI Chat completion response object
print(content)
```

```python
model_name = "claude-sonnet-4-5"
response = claude.messages.create(model=model_name, messages=message, max_tokens=1000)
answer = response.content[0].text
```
